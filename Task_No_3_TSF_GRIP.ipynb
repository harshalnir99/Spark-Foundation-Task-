```python
# Data Visulaization
```

```python
#categories wise Sales and Profit

dataset.groupby('Category')['Profit','Sales'].sum().plot.bar(figsize=(12,6))

plt.ylabel('Profit & sales')
plt.title("Profit and Sales by category")
plt.legend(loc = 2)
plt.show()
```

```python
# In Category wise Sales & profit, I Observed Technology makes more Profit & Furniture Make less Profit
```

```python
#Region wise Sales and Profit

dataset.groupby('Region')["Profit","Sales"].sum().plot.bar(figsize=(12,6))

plt.ylabel('Profit/Sales')
plt.title("Profit and Sales by Region")
plt.legend(loc = 2)
plt.show()
```

```python
# In Region wise Sales & profit, I Observed West region makes more Profit & Central region Make less Profit
```

```python
#Sub-Categories wise Profit

plt.figure(figsize=(10,4))

dataset.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False).plot.bar(color="green", edgecolor="black")

plt.show()
```

```python
# In Sub-categories wise profit, I Observed Copies,Phones,Accessories,paper, Binder makes more Profit & Tables,Bookcases 
#& supplies Make losses
```

```python
#Ship Mode wise Sales and Profit

dataset.groupby('Ship Mode')["Profit","Sales"].sum().plot.bar(figsize=(12,6))

plt.ylabel('Profit/Sales')
plt.title("Profit and Sales by Ship Mode")
plt.legend(loc = 2)
plt.show()
```

```python
#In ship-mode wise Sales & profit, I Observed Standard Class mode makes more Profit & Same day mode Make less Profit
```

```python
#Segment wise Sales and Profit

dataset.groupby('Segment')["Profit","Sales"].sum().plot.bar(figsize=(12,6))

plt.ylabel('Profit/Sales')
plt.title("Profit and Sales by Segemnt")
plt.legend(loc = 1)
plt.show()
```

```python
#In Segment wise Sales & profit, I Observed Consumer makes more Profit & Home Office Make less Profit
```

```python
# Top 5 Profit Cities

plt.figure(figsize=(12,6))
dataset.groupby(['City'])['Profit'].sum().sort_values(ascending=False).head().plot.bar(color=['Green'])
plt.xlabel("City")
plt.ylabel("Profit")
plt.title("City wise Profit",fontdict={'color':'black','fontsize':20})
plt.show()
```

```python
# in City wise top 5 profit cities, New york, loss Angeles,Seattle, San Francisco, Detroit
```

```python
# Loss making 5 Cities

plt.figure(figsize=(12,6))
dataset.groupby(['City'])['Profit'].sum().sort_values(ascending=False).tail().plot.bar(color=['red'])
plt.xlabel("City")
plt.ylabel("Profit")
plt.title("City wise loss",fontdict={'color':'black','fontsize':20})
plt.show()
```

```python
# in City wise 5 loss cities, Chicago, Lancaster, San Antonio, Houstain, Philadelphia
```

```python
#Comparison Plot

figsize=(15,10)
sns.pairplot(dataset,hue='Sub-Category')
plt.show()
```

```python
#  Profit Vs Sales plot

dataset.plot(x='Sales', y='Profit', style='o')  
plt.title('Sales vs Profit')  
plt.xlabel('Sales')  
plt.ylabel('Profit')  
plt.show()
```
